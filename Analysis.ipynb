```python
import numpy as np
import torch
import pickle
import matplotlib.pyplot as plt
from datetime import datetime

import matplotlib.ticker as ticker
```

# Experiment 2

```python
path_SNR20 = './results/exprm_2/summary_res_SNR20.pkl'
path_SNR40 = './results/exprm_2/summary_res_SNR40.pkl'
path_SNR80 = './results/exprm_2/summary_res_SNR80.pkl'
```

```python
N0s = [5, 10, 15, 20, 25, 30, 35, 40, 45]
m = 300
ratios = np.array(N0s)/m
```

#### Load SNR = 40

```python
with open(path_SNR40, 'rb') as handle:
    summary_res_SNR40 = pickle.load(handle)
```

```python
relError_y_mean_SNR40 = []
relError_y_std_SNR40 = []
relError_x_mean_SNR40 = []
relError_x_std_SNR40 = []

for res in summary_res_SNR40:
    relError_y_mean_SNR40.append(res['relError_y_mean'])
    relError_y_std_SNR40.append(res['relError_y_std'])
    relError_x_mean_SNR40.append(res['relError_x_mean'])
    relError_x_std_SNR40.append(res['relError_x_std'])
```

#### Load SNR = 20

```python
with open(path_SNR20, 'rb') as handle:
    summary_res_SNR20 = pickle.load(handle)
```

```python
relError_y_mean_SNR20 = []
relError_y_std_SNR20 = []
relError_x_mean_SNR20 = []
relError_x_std_SNR20 = []

for res in summary_res_SNR20:
    relError_y_mean_SNR20.append(res['relError_y_mean'])
    relError_y_std_SNR20.append(res['relError_y_std'])
    relError_x_mean_SNR20.append(res['relError_x_mean'])
    relError_x_std_SNR20.append(res['relError_x_std'])
```

#### Load SNR = 80

```python
with open(path_SNR80, 'rb') as handle:
    summary_res_SNR80 = pickle.load(handle)
```

```python
relError_y_mean_SNR80 = []
relError_y_std_SNR80 = []
relError_x_mean_SNR80 = []
relError_x_std_SNR80 = []

for res in summary_res_SNR80:
    relError_y_mean_SNR80.append(res['relError_y_mean'])
    relError_y_std_SNR80.append(res['relError_y_std'])
    relError_x_mean_SNR80.append(res['relError_x_mean'])
    relError_x_std_SNR80.append(res['relError_y_std'])
```

### Plots

#### Error Plots

```python
fig, ax = plt.subplots(1,3)
fig.set_size_inches(15, 8)
ax[0].errorbar(N0s,relError_y_mean_SNR20,yerr=relError_y_std_SNR20, marker='o',markersize=8,capsize=5,color='blue',label='SNR=20')
ax[0].set_xlabel(r'$\frac{k}{m}$',fontsize=20)
ax[0].set_ylabel('Reconstruction Error',fontsize=15)
ax[0].ticklabel_format(style='sci', axis='y')
ax[0].yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.0e'))
ax[0].grid()
ax[1].errorbar(N0s,relError_y_mean_SNR40,yerr=relError_y_std_SNR40, marker='o',markersize=8,capsize=5,color='black',label='SNR=40')
ax[1].set_xlabel(r'$\frac{k}{m}$',fontsize=20)
ax[1].set_ylabel('Reconstruction Error',fontsize=15)
ax[1].yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.0e'))
ax[1].grid()
ax[2].errorbar(N0s,relError_y_mean_SNR80,yerr=relError_y_std_SNR80, marker='o',markersize=8,capsize=5,color='magenta',label='SNR=80')
ax[2].set_xlabel(r'$\frac{k}{m}$',fontsize=20)
ax[2].set_ylabel('Reconstruction Error',fontsize=15)
ax[2].yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.0e'))
ax[2].grid()
fig.legend(bbox_to_anchor=(.02, -.1, 1., .102), loc='center',
           ncol=3, borderaxespad=0.,fontsize=15)
fig.tight_layout()
plt.savefig('./results/exprm_2/reconstruction_error_bar.pdf', bbox_inches='tight', dpi=250)
plt.show()
```

```python
fig, ax = plt.subplots(1,3)
fig.set_size_inches(15, 8)
ax[0].errorbar(N0s,relError_x_mean_SNR20,yerr=relError_x_std_SNR20, marker='o',markersize=8,color='blue',label='SNR=20')
ax[0].set_xlabel('k',fontsize=15)
ax[0].set_ylabel('Reconstruction Error',fontsize=15)
ax[0].set_title('Noisy Compressed Sensing')
ax[0].grid()
ax[1].errorbar(N0s,relError_x_mean_SNR40,yerr=relError_x_std_SNR40, marker='o',markersize=8,color='black',label='SNR=40')
ax[1].set_xlabel('k',fontsize=15)
ax[1].set_ylabel('Reconstruction Error',fontsize=15)
ax[1].set_title('Noisy Compressed Sensing')
ax[1].grid()
ax[2].errorbar(N0s,relError_x_mean_SNR80,yerr=relError_x_std_SNR80, marker='o',markersize=8,color='magenta',label='SNR=80')
ax[2].set_xlabel('k',fontsize=15)
ax[2].set_ylabel('Reconstruction Error',fontsize=15)
ax[2].set_title('Noisy Compressed Sensing')
ax[2].grid()
fig.legend(bbox_to_anchor=(.02, -.1, 1., .102), loc='center',
           ncol=3, borderaxespad=0.,fontsize=15)
fig.tight_layout()
# plt.savefig('nontun_v_tun.png', bbox_inches='tight',dpi=150)
plt.show()
```

#### Plots

```python
fig, ax = plt.subplots(1,3)
fig.set_size_inches(15, 8)
ax[0].plot(ratios,relError_y_mean_SNR20, marker='o',markersize=8,color='blue',label='SNR=20')
ax[0].set_xlabel(r'$\frac{k}{m}$',fontsize=20)
ax[0].set_ylabel('Reconstruction Error',fontsize=15)
ax[0].yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.0e'))
ax[0].grid()
ax[1].plot(ratios,relError_y_mean_SNR40, marker='o',markersize=8,color='black',label='SNR=40')
ax[1].set_xlabel(r'$\frac{k}{m}$',fontsize=20)
ax[1].set_ylabel('Reconstruction Error',fontsize=15)
ax[1].yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.0e'))
ax[1].grid()
ax[2].plot(ratios,relError_y_mean_SNR80, marker='o',markersize=8,color='magenta',label='SNR=80')
ax[2].set_xlabel(r'$\frac{k}{m}$',fontsize=20)
ax[2].set_ylabel('Reconstruction Error',fontsize=15)
ax[2].yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.0e'))
ax[2].grid()
fig.legend(bbox_to_anchor=(.02, -.1, 1., .102), loc='center',
           ncol=3, borderaxespad=0.,fontsize=15)
fig.tight_layout()
plt.savefig('./results/exprm_2/reconstruction_error.pdf', bbox_inches='tight', dpi=250)
plt.show()
```

```python
fig, ax = plt.subplots(1,3)
fig.set_size_inches(15, 8)
ax[0].plot(ratios,relError_x_mean_SNR20, marker='o',markersize=8,color='blue',label='SNR=20')
ax[0].set_xlabel(r'$\frac{k}{m}$',fontsize=20)
ax[0].set_ylabel('Recovery Error',fontsize=15)
ax[0].yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.0e'))
ax[0].grid()
ax[1].plot(ratios,relError_x_mean_SNR40, marker='o',markersize=8,color='black',label='SNR=40')
ax[1].set_xlabel(r'$\frac{k}{m}$',fontsize=20)
ax[1].set_ylabel('Recovery Error',fontsize=15)
ax[1].yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.0e'))
ax[1].grid()
ax[2].plot(ratios,relError_x_mean_SNR80, marker='o',markersize=8,color='magenta',label='SNR=80')
ax[2].set_xlabel(r'$\frac{k}{m}$',fontsize=20)
ax[2].set_ylabel('Recovery Error',fontsize=15)
ax[2].yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.0e'))
ax[2].grid()
fig.legend(bbox_to_anchor=(.02, -.1, 1., .102), loc='center',
           ncol=3, borderaxespad=0.,fontsize=15)
fig.tight_layout()
fig.set_tight_layout(True)
plt.savefig('./results/exprm_2/recovery_error.pdf', bbox_inches='tight', dpi=250)
plt.show()
```

# Experiment 1

```python
summary_res_2Layers_path = './results/exprm_1/summary_res_2Layers.pkl'
summary_res_3Layers_path = './results/exprm_1/summary_res_3Layers.pkl'
summary_res_2Layers_k120_path = './results/exprm_1/summary_res_2Layers_k120.pkl'
summary_res_2Layers_k135_path = './results/exprm_1/summary_res_2Layers_k135.pkl'
```

```python
N0s_3L = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 70, 80, 90, 100, 150]
N0s_2L = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 70, 80, 90, 100, 120, 135, 150]
m = 300
ratios = np.array(N0s)/m
```

#### Load 2 Layers

```python
with open(summary_res_2Layers_path, 'rb') as handle:
    summary_res_2Layers = pickle.load(handle)

with open(summary_res_2Layers_k120_path, 'rb') as handle:
    summary_res_2Layers_k120 = pickle.load(handle)

with open(summary_res_2Layers_k135_path, 'rb') as handle:
    summary_res_2Layers_k135 = pickle.load(handle)
```

```python
summary_res_2Layers.insert(16, summary_res_2Layers_k120)
summary_res_2Layers.insert(17, summary_res_2Layers_k135)
```

```python
relError_y_mean_2L = []
relError_y_std_2L = []
relError_x_mean_2L = []
relError_x_std_2L = []
prob_recovery_x_2L = []

for res in summary_res_2Layers:
    relError_y_mean_2L.append(res['relError_y_mean'])
    relError_y_std_2L.append(res['relError_y_std'])
    relError_x_mean_2L.append(res['relError_y_mean'])
    relError_x_std_2L.append(res['relError_y_std'])
    prob_recovery_x_2L.append(res['prob_recovery_x'])
```

#### Load 3 Layers

```python
with open(summary_res_3Layers_path, 'rb') as handle:
    summary_res_3Layers = pickle.load(handle)
```

```python
relError_y_mean_3L = []
relError_y_std_3L = []
relError_x_mean_3L = []
relError_x_std_3L = []
prob_recovery_x_3L = []

for res in summary_res_3Layers:
    relError_y_mean_3L.append(res['relError_y_mean'])
    relError_y_std_3L.append(res['relError_y_std'])
    relError_x_mean_3L.append(res['relError_y_mean'])
    relError_x_std_3L.append(res['relError_y_std'])
    prob_recovery_x_3L.append(res['prob_recovery_x'])
```

### Plots

```python
path = './results/experiment_1_v2/'

plt.plot(N0s_2L, prob_recovery_x_2L, 'b-*', label = '2 Layers')
plt.plot(N0s_3L, prob_recovery_x_3L, 'm-*', label = '3 Layers')
plt.legend(loc = 'best')
plt.xlabel(r'$k$', fontsize=15)
plt.grid()
plt.ylabel('Probability of Success', fontsize = 16)
plt.savefig('./results/exprm_1/prob_succ.pdf', bbox_inches='tight', dpi=250)
```
